```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def freq(n):
    return 440.0 * 2 ** (n/12)
```

```python
DB = [-12, 7, 0, 12, 12+7, 2*12, 2*12+4, 2*12+7, 3*12]
```

```python
[freq(n) for n in DB]
```

```python
FS = 44100.0
t = np.arange(0, 0.02, 1/FS)
```

```python
x = sum([np.sin(2 * np.pi * freq(n) * t) for n in DB])
```

```python
len(x)
```

```python
plt.plot(t, x)
```
